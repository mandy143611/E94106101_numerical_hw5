```python
import numpy as np
import matplotlib.pyplot as plt

# 定義 f(t, y)
def f(t, y):
    return 1 + (y / t) + (y / t)**2

# 解析解
def exact_solution(t):
    return t * np.tan(np.log(t))

# 初始條件
t0, y0 = 1.0, 0.0
h = 0.1
N = int((2.0 - 1.0) / h)

# 建立 Euler 解
t_vals = [t0]
y_vals = [y0]

t = t0
y = y0

for i in range(N):
    y = y + h * f(t, y)
    t = t + h
    t_vals.append(t)
    y_vals.append(y)

# 計算解析解
t_exact = np.linspace(1, 2, 100)
y_exact = exact_solution(t_exact)

# 比較數值與解析值
print("t\tEuler y\t\tExact y\t\tError")
for i in range(len(t_vals)):
    exact_y = exact_solution(t_vals[i])
    error = abs(y_vals[i] - exact_y)
    print(f"{t_vals[i]:.1f}\t{y_vals[i]:.6f}\t{exact_y:.6f}\t{error:.6f}")

# 繪圖
plt.plot(t_exact, y_exact, label='Exact Solution', color='blue')
plt.plot(t_vals, y_vals, 'o--', label='Euler Approximation', color='red')
plt.title("Euler's Method vs Exact Solution")
plt.xlabel("t")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
